## Knapsack Problem
The Knapsack Problem is a classic problem in combinatorial optimization. You are given a set of items, each with a weight and a value, and a knapsack that can carry a maximum weight. The challenge is to determine the number of each item to include in the knapsack such that the total weight is less than or equal to the knapsack's capacity and the total value is as large as possible.

### 1. Define the Knapsack Problem

Given a list of items where each item has a weight and a value, and a maximum weight capacity of the knapsack, find the combination of items that maximizes the total value without exceeding the knapsack's weight limit.

```python
class Item:
    def __init__(self, value, weight):
        self.value = value
        self.weight = weight

class KnapsackProblem:
    def __init__(self, items, capacity):
        self.items = items
        self.capacity = capacity
        self.initial = [0] * len(items)  # Initial state with no items in the knapsack

    def value(self, state):
        total_value = sum(item.value for item, in_knapsack in zip(self.items, state) if in_knapsack)
        total_weight = sum(item.weight for item, in_knapsack in zip(self.items, state) if in_knapsack)
        if total_weight > self.capacity:
            return 0  # Penalize for exceeding the weight limit
        else:
            return total_value

    def expand(self, state):
        neighbors = []
        for i in range(len(state)):
            neighbor = state[:]
            neighbor[i] = 1 - neighbor[i]  # Flip the bit to generate a neighbor
            neighbors.append(neighbor)
        return neighbors

class Node:
    def __init__(self, state):
        self.state = state

    def expand(self, problem):
        return [Node(neighbor) for neighbor in problem.expand(self.state)]

def argmax_random_tie(seq, key):
    return max(random.sample(seq, len(seq)), key=key)
```

## Erratic Vacuum Problem

Consider a world with two squares, A and B, each of which can be either clean or dirty. The vacuum cleaner can be in either square and perform the following actions:

    Move left (from B to A)
    Move right (from A to B)
    Suck dirt

The "Suck" action has uncertain outcomes:

    When applied to a dirty square, it cleans the square and sometimes cleans the adjacent square too.
    When applied to a clean square, it sometimes deposits dirt on the floor.

Represent the state with a tuple (vacuum location, A status, B status). For example, (A, dirty, clean) means the vacuum is at A, square A is dirty, and square B is clean.

```python
class ErraticVacuumProblem:
    def __init__(self, initial):
        self.initial = initial  # Initial state format: (vacuum_location, A_status, B_status)

    def goal_test(self, state):
        # Goal is both A and B are clean
        return state[1] == 'clean' and state[2] == 'clean'

    def actions(self, state):
        actions_list = ['Suck']
        if state[0] == 'A':  # Can only move right if at A
            actions_list.append('Right')
        elif state[0] == 'B':  # Can only move left if at B
            actions_list.append('Left')
        return actions_list

    def result(self, state, action):
        # Results of performing an action from a state
        if action == 'Suck':
            # Suck action can have two different outcomes
            if state[0] == 'A':  # If vacuum is at A
                if state[1] == 'dirty':
                    # Suck cleans A for sure, might clean B
                    return [('A', 'clean', state[2]), ('A', 'clean', 'clean')]
                else:
                    # If A is clean, suck might make A dirty
                    return [('A', 'dirty', state[2])]
            else:  # If vacuum is at B
                if state[2] == 'dirty':
                    # Suck cleans B for sure, might clean A
                    return [('B', state[1], 'clean'), ('B', 'clean', 'clean')]
                else:
                    # If B is clean, suck might make B dirty
                    return [('B', state[1], 'dirty')]
        elif action == 'Left' and state[0] == 'B':
            # Move left (from B to A)
            return [('A', state[1], state[2])]
        elif action == 'Right' and state[0] == 'A':
            # Move right (from A to B)
            return [('B', state[1], state[2])]
        else:
            # If the action is not applicable, return an empty list of results
            return []
```
