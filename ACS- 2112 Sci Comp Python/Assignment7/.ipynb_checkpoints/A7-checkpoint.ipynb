# NumPy Array Attributes

# Generating random numbers

One dimensional array

```python
import numpy as np
rng = np.random.default_rng(seed = 1701)
n1 = rng.integers(10, size = 10)
print(n1)
```

Two dimensional array

```python
n2 = rng.integers(10 , size = (5 , 7))
print(n2)
```

Three dimensional array

```python
n3 = rng.integers(10 , size = (4 , 4 , 5))
print(n3)
```

Sum of all the random numbers

```python
rng = np.random.default_rng()
N = rng.random(200)
np.sum(N)
```

```python
sum(N)
```

# Reshaping of arrays

```python
r1 = np.arange(1, 10).reshape(3, 3)
print(r1)
```

# Minimum and Maximum

# Multidimensional Aggregates

```python
M = rng.integers(0 , 10 , (4 , 6))
print(M)
```

```python
M.sum() #Sum all the integers
```

```python
M.min(axis=0)
```

```python
M.max(axis=0)
```

# Broadcasting

```python
a = np.array([0, 1, 4])
b = np.array([2, 4, 5])
a + b
```

```python
a + 5     #add 5 in each element of the array
```

Write the arrays in different axis

```python
a = np.arange(5)
b = np.arange(5)[:, np.newaxis]
print(a)
print(b)
```

adding two dimensional array to one dimensional array

```python
T = np.ones((2, 3))
print(T)
```

```python
p = np.arange(3)
print(p)
```

```python
T + p
```

# Centering an Array

```python
n1mean = n1.mean(0)
print(n1mean)
```

```python
n1_centered = n1 - n1mean
print(n1_centered)
```

```python
n1_centered.mean(0)
```
