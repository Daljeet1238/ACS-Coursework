```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
census2001 = pd.read_csv("Census_2001_-__Citizenship___Immigration_20240412.csv")
census2006 = pd.read_csv("Census_2006_-__Citizenship___Immigration_20240412.csv")
```

```python
#It plots the first 50 data points from the 'Citizen Canadian' 
#column of the DataFrame census2001.
sns.set()
sns.lineplot(data=census2001['Citizen Canadian'].head(50), marker='o', linestyle='-')

plt.xlabel('Data Number')
plt.ylabel('No. of Citizen')
plt.title('Line Graph')
plt.show()
```

```python
#This line of code concatenates the first 30 data points from the 'Citizen Canadian' 
#columns of the census2006 and census2001
combined_data = pd.concat([census2006['Citizen Canadian'].head(30), census2001['Citizen Canadian'].head(30)], axis=1)
combined_data.columns = ['2006', '2001']

sns.set()
plt.figure(figsize=(10, 6))
sns.lineplot(data=combined_data, marker='o', linestyle='-')

plt.xlabel('Data Number')
plt.ylabel('No. of Citizens')
plt.title('Comparison of Citizen Canadian between 2001 and 2006')
plt.show()
```

```python
cnc2001 = census2001["Citizen Non-Canadian"].tail(20)
fig, ax = plt.subplots()

ax.bar(cnc2001.index, cnc2001)
plt.show()
```

```python
cnc2006 = census2006["Citizen Non-Canadian"].tail(20)
fig, ax = plt.subplots()

ax.bar(cnc2006.index, cnc2006)
plt.show()
```

```python
cnc2001 = census2001["Citizen Non-Canadian"].tail(15)
cnc2006 = census2006["Citizen Non-Canadian"].tail(15)
```

```python
#This code creates a grouped bar plot using Matplotlib, comparing the 
#'Citizen Non-Canadian' data from Census 2001 and Census 2006.
fig, ax = plt.subplots()
width = 0.35
x = range(len(cnc2001))

ax.bar(x, cnc2001, width, label='Census 2001')
ax.bar([i + width for i in x], cnc2006, width, label='Census 2006')

ax.set_xlabel('Index')
ax.set_ylabel('Citizen Non-Canadian')
ax.set_title('Comparison of Citizen Non-Canadian in Census 2001 and 2006')

ax.set_xticks([i + width / 2 for i in x])
ax.set_xticklabels(cnc2001.index)
plt.xticks(rotation=45, ha='right')

ax.legend()
plt.show()
```

```python
#This code creates a histogram of the data in the variable cnc2001
plt.figure(figsize=(8, 6))
sns.histplot(cnc2001, kde=False, stat='density', alpha=0.5)
plt.title('Histogram of Citizen Non-Canadian')

plt.tight_layout()
plt.show()
```

```python
#This code creates a histogram of the data in the variable cnc2006
plt.figure(figsize=(8, 6))
sns.histplot(cnc2006, kde=False, stat='density', alpha=0.5)
plt.title('Histogram of Citizen Non-Canadian')

plt.tight_layout()
plt.show()
```

```python
#This code generates side-by-side histograms comparing the distributions of 
#non-Canadian citizens in 2001 and 2006
plt.figure(figsize=(8, 6))

sns.histplot(cnc2001, kde=False, stat='density', alpha=0.5, color='blue', label='2001')
sns.histplot(cnc2006, kde=False, stat='density', alpha=0.5, color='orange', label='2006')

plt.title('Histogram of Citizen Non-Canadian')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#This code creates a kernel density plot (KDE) of the distribution of 
#non-Canadian citizens in 2006
plt.figure(figsize=(8, 6))
sns.kdeplot(cnc2006, shade=True)
plt.title('Histogram of Citizen Non-Canadian')

plt.tight_layout()
plt.show()
```

```python
#This code creates a kernel density plot (KDE) of the distribution of 
#non-Canadian citizens in 2001
plt.figure(figsize=(8, 6))
sns.kdeplot(cnc2001, shade=True)
plt.title('Histogram of Citizen Non-Canadian')

plt.tight_layout()
plt.show()
```

```python
#it compares the upper data.
plt.figure(figsize=(8, 6))

sns.kdeplot(cnc2001, shade=True, label='2001')
sns.kdeplot(cnc2006, shade=True, label='2006')

plt.title('Histogram of Citizen Non-Canadian')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#This code creates side-by-side histograms and kernel density plots for the 
#distributions of non-Canadian citizens in 2001 and 2006
plt.figure(figsize=(10, 6))

sns.histplot(cnc2001, kde=False, stat='density', alpha=0.5, color='blue', label='2001')
sns.kdeplot(cnc2001, shade=True, color='blue', label='2001')

sns.histplot(cnc2006, kde=False, stat='density', alpha=0.5, color='orange', label='2006')
sns.kdeplot(cnc2006, shade=True, color='orange', label='2006')

plt.title('Histogram and KDE of Citizen Non-Canadian')
plt.legend()
plt.tight_layout()
plt.show()
```
