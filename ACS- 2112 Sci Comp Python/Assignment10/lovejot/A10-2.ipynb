```python
target_value = 'City Transportation'

counts_dict = {}

for index, row in data1.iterrows():
    column_value = row['Street Qualifier']
    
    if target_value in str(row.values):
        counts_dict[column_value] = counts_dict.get(column_value, 0) + 1

for value, count in counts_dict.items():
    print("Number of times '{}' appears in the same row with '{}'".format(target_value, value, count))
```

```python
target_value = 'City Transportation'

counts_dict = {}

for index, row in data1.iterrows():
    column_value = row['Street Qualifier']
    
    if target_value in str(row.values):
        counts_dict[column_value] = counts_dict.get(column_value, 0) + 1

sorted_counts = sorted(counts_dict.items(), key=lambda x: x[1], reverse=True)

top_10_counts = sorted_counts[:10]

for value, count in top_10_counts:
    print("Number of times '{}' appears in the same row with '{}': {}".format(target_value, value, count))
```

```python
values = [x[0] for x in top_10_counts]
counts = [x[1] for x in top_10_counts]

mplt.figure(figsize=(10, 6))
mplt.bar(values, counts, color='skyblue')
mplt.xlabel('Street Qualifier')
mplt.ylabel('Frequency')
mplt.title("Top 10 Street Qualifiers with '{}'".format(target_value))
mplt.xticks(rotation=45, ha='right')
mplt.tight_layout()
mplt.show()
```
