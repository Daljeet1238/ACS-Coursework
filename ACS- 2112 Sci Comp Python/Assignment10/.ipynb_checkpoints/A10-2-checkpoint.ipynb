```python
# Count the occurrences where 'Brian Mayes' and 'CWA' appear in the same row but in different columns 
# in the DataFrame cme
import pandas as pd

cme = pd.read_csv('Council_Member_Expenses_20240403.csv')


value_a = 'Brian Mayes'
value_b = 'CWA'

count = 0

for index, row in cme.iterrows():
    if (value_a in str(row.values)) and (value_b in str(row.values)):
        count += 1

print("Number of times {} and {} occur in the same row of different columns: {}".format(value_a, value_b, count))
```

```python
value_a = 'CWA'

counts_dict = {}

for index, row in cme.iterrows():
    # Get the value from "Council Mmember" column 
    column_one_value = row['Council Member']
    
    if value_a in str(row.values):
        counts_dict[column_one_value] = counts_dict.get(column_one_value, 0) + 1

# Print the counts for each value from "Council Mmember" column
for value, count in counts_dict.items():
    print("Number of times '{}' appears in the same row with '{}'".format(value_a, value, count))
```

This code generates a DataFrame containing the counts of occurrences of each combination of "Council Member" and "Expense Type" for the top 5 most frequent "Council Members" in the dataset.

```python
top_council_members = cme['Council Member'].value_counts().head(5).index

cme_top = cme[cme['Council Member'].isin(top_council_members)]

# Define the unique values from "Council Member" and "Expense Type" columns for the top 5 Council Members
council_member_values = cme_top['Council Member'].astype(str).unique()
expense_type_values = cme_top['Expense Type'].astype(str).unique()
counts_dict = {}

for index, row in cme_top.iterrows():
    council_member_value = str(row['Council Member'])
    
    for expense_type_value in expense_type_values:
        if expense_type_value in str(row.values):
            counts_dict[(council_member_value, expense_type_value)] = counts_dict.get((council_member_value, expense_type_value), 0) + 1

counts_df = pd.DataFrame.from_dict(counts_dict, orient='index', columns=['Count'])

# Reset index to create separate columns for "Council Member" and "Expense Type"
counts_df.reset_index(inplace=True)
counts_df.rename(columns={'index': 'Combination'}, inplace=True)
counts_df[['Council Member', 'Expense Type']] = pd.DataFrame(counts_df['Combination'].tolist(), index=counts_df.index)


```

```python
mplt.figure(figsize=(10, 6))
for council_member in top_council_members:
    data = counts_df[counts_df['Council Member'] == council_member]
    mplt.bar(data['Expense Type'], data['Count'], label=council_member)

mplt.title('Counts of Expense Types for Top 5 Council Members')
mplt.xticks(rotation=45, ha='right')
mplt.legend()
mplt.show()
```
